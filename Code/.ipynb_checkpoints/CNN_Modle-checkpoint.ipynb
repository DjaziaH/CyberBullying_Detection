```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
import pandas as pd
from tensorflow.keras.layers import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from gensim.models import Word2Vec
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../Data/PreProcessedYouTubeDataFileAndAnnotated.csv')
```

```python
data["num_words"] = data["text"].apply(lambda x : len(str(x).split()))
sent_length = data["num_words"].max()
```

```python
X = data['text'].tolist()
print(X)
```

```python
type(X)
```

```python
Y = np.where(data["offensive/non offensive"] == "p",1,0)
print(Y)
```

```python
type(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.35)
```

```python
### Vocabulary size
voc_size=10000
```

```python
onehot_train=[one_hot(words,voc_size)for words in X_train] 
onehot_test=[one_hot(words,voc_size)for words in X_test] 
print(onehot_train)
print('-----------------------')
print(onehot_test)
```

```python
embedded_train=pad_sequences(onehot_train,padding='pre',maxlen=sent_length)
embedded_test =pad_sequences(onehot_test ,padding='pre',maxlen=sent_length)
print(embedded_train.shape)
print(embedded_test.shape)
```

```python
word_dim=300
```

```python
model=Sequential()
model.add(Embedding(voc_size,word_dim,input_length=sent_length))
model.add(Conv1D(40,5,activation="relu"))
model.add(MaxPooling1D(2,2))
model.add(Conv1D(20,2,activation="relu"))
model.add(MaxPooling1D(2,2))
model.add(Dense(10,activation="relu"))
model.add(Flatten())
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy','Recall','Precision','TruePositives','TrueNegatives','FalsePositives','FalseNegatives'])
```

```python
model.summary()
```

```python
model.fit(embedded_train, y_train, epochs = 100, batch_size = 128, validation_split=0.2)
```

```python
predictions = model.predict(embedded_test)
print(predictions)
```

```python
predictions = [predictions[i][0] for i in range(len(predictions))]
predict_result = [round(num) for num in predictions]
```

```python
print(sum(predict_result))
print(sum(y_test))
```

```python
print(classification_report(y_test, predict_result))
```
